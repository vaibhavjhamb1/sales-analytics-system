```python
import requests

def fetch_product_info(product_id):
    """
    Fetches real-time product information from an external API.
    """
    # Example placeholder URL - replace with the actual API endpoint provided
    api_url = f"https://api.example.com/products/{product_id}"
    
    try:
        response = requests.get(api_url, timeout=5)
        
        # Check if the request was successful
        response.raise_for_status()
        
        # Parse and return JSON data
        return response.json()
        
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred for Product {product_id}: {http_err}")
    except requests.exceptions.ConnectionError:
        print(f"Error: Could not connect to the API for Product {product_id}.")
    except Exception as err:
        print(f"An unexpected error occurred: {err}")
    
    return None
```
